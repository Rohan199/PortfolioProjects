# Movies Data Correlation

First lets import the libraries as well as load the data.

```python
#import libraries
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #adjusts the configuration of the plots we will create

#load data

df= pd.read_csv(r"C:\Users\rohan\Desktop\Project 1\movies.csv\movies.csv")
```

```python
#lets preview the data

df.head()
```

Now, lets explore the data.

```python
#checking for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

So far, we can see that there are missing values in the table. We will be dropping the rows with missing values for the sake of simplicity even though while dealing with real world data it might be a less desirable option.

```python
df = df.dropna()  #drops rows with null value
```

```python
# dataypes for the columns

df.dtypes
```

```python
# adjusting datatypes for budget and gross to int64

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')

df
```

Since, the year column and the released column not always reflect the same year, let's extract the year from the released column.

```python

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)    #extracts the YYYY format from the relased column using str.extract method

```

```python
df.head()
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False) 
```

```python
pd.set_option('display.max_rows', None) #to be able to navigate the whole data.
```

Now that we have a cleaner data, lets start to find correlation.

```python
# Hypothesis: Budget have high correlation, company have high correlation

#Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')

plt.xlabel('Budget for the film')
plt.ylabel('Gross Earnings')

plt.show()
```

```python
#plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color':'orange'}, line_kws={'color':'blue'})
```

Now that we see a correlation, lets dig deeper.

```python
df.corr(numeric_only=True)  #Pearson correlation
```

We see that 'votes' and 'budget' have strong correlation with the gross earning.

```python
#correlation matrix

correlation_matrix= df.corr(numeric_only=True)

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

Now, let's look at companies and their correlation to gross earnings.

```python
#visualizing the gross revenue vs company for the top 50 grossing companies
plt.figure(figsize=(12, 6))
sns.boxplot(x='company', y='gross', data=df.head(50))
plt.xticks(rotation=45)
plt.title('Box Plot of Gross Revenue by Company for the top 50 highest grossing movies')
plt.show()
```

We see that some companies produce more highest grossing movies compared to others.

```python
from scipy.stats import f_oneway
```

```python
#perform ANOVA
anova_result = f_oneway(*[df['gross'][df['company'] == company] for company in df['company'].unique()])

#display ANOVA result
print("ANOVA p-value:", anova_result.pvalue)

#post-hoc Testing (Tukey's HSD)
from statsmodels.stats.multicomp import pairwise_tukeyhsd

#perform Tukey's HSD test
tukey_result = pairwise_tukeyhsd(df['gross'].head(50), df['company'].head(50))

#display Tukey's HSD result
print(tukey_result)
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()
```

```python
#correlation matrix
correlation_matrix= df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```
